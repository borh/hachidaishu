```python
import csv

def read_db():
    with open('hachidai.db', newline='') as f:
        reader = csv.reader(f, delimiter=' ')
        for row in reader:
            id, token_type, bg_id, pos_chasen, surface, lemma, lemma_reading, orthbase, orthbase_reading, _ = row
            anthology, poem, serial = id.split(':')
            token = {
                'token_type': token_type, 
                'bg_id': bg_id,
                'pos_chasen': pos_chasen,
                'surface': surface, 
                'lemma': lemma, 
                'lemma_reading': lemma_reading, 
                'orthbase': orthbase,
                'orthbase_reading': orthbase_reading
            }
            yield {
                'anthology': int(anthology),
                'poem': int(poem), 
                'serial': int(serial),
                'token': {token_type: token}
                }




list(read_db())[0:5]
```

```python
def tokenize(db):
    tokens = []
    serial = None
    for entry in db:
        
        if serial == entry['serial']:
            tokens[-1]['token'] = {**tokens[-1]['token'], **entry['token']}
        else:
            tokens.append(entry)
        
        serial = entry['serial']
    return tokens

tokenize(read_db())[:10]
```
